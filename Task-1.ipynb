#### Importing necessary packages

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

#### Data Upload

```python
df=pd.read_csv("iris.data",names=["Sepal Length","Sepal Width","Petal Length","Petal Width","Class"])
pd.set_option("display.max_rows",None,"display.max_columns",None,"display.precision",3)
print(df)
```

```python
print(df.describe())
```

```python
print(df.head())
```

```python
print(df.tail())
```

#### Checking for null values

```python
print(df.isnull().sum())
```

```python
print(df["Class"].describe(include="all"))
print(set(df["Class"]))
count = df.groupby(["Class"]).size()
print(count)
```

#### Pie-chart of the various classes

```python
my_labels = (['Iris-versicolor', 'Iris-setosa', 'Iris-virginica'])
my_explode = (0.05, 0.05, 0.05)
_,_,autotexts=plt.pie(df.groupby(['Class']).size(),labels=my_labels,autopct='%1.1f%%',startangle=-15, shadow = True, explode=my_explode)
for autotext in autotexts:
    autotext.set_color('black')
plt.title('Distribution Based on Classes'.upper(),fontsize=20,color='black',fontstyle='oblique',loc='center')
plt.axis('equal')
plt.subplots_adjust(bottom=0, top=1, left=1, right=3)
plt.show()
```

#### Histogram

```python
df.hist()
plt.show()
```

#### Density Plot

```python
df.plot(kind ='density',subplots = True, layout =(3,3),sharex = False)
plt.show()
```

#### Box-Plot

```python
df.plot(kind ='box',subplots = True, layout =(2,5),sharex = False)
plt.show()
```

#### Violin-plot

```python
fig, axes = plt.subplots(2,2,figsize=(18, 10))
sns.violinplot(ax=axes[0,0],x="Class",y="Sepal Length",data=df)
sns.violinplot(ax=axes[0,1],x="Class",y="Sepal Width",data=df)
sns.violinplot(ax=axes[1,0],x="Class",y="Petal Length",data=df)
sns.violinplot(ax=axes[1,1],x="Class",y="Petal Width",data=df)
plt.show()
```
